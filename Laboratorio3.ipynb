#### 2). Cree un array cuyos elementos sean los enteros pares en [1,100] y en orden decreciente. Muestre los 10 últimos por pantalla


```python
import numpy as np

# Creamos un array con elementos enteros de 1 a 100
array_enteros_all = np.arange(1, 101)

# Filtramos los enteros pares
array_pares = array_enteros_all[array_enteros_all % 2 == 0]

print('array_pares:', array_pares)
```

```python
# Ordenando de forma decreciente
array_pares_dec = np.sort(array_pares)[::-1]

# Top 10 de los ultimos enteros pares:
top_10 = array_pares_dec[-10:]

print("Los últimos 10 enteros pares en el rango [1, 100] en orden decreciente son:")
print(top_10)
```

#### 4) Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], obtenga la media aritmética de la diferencia (a-b).

```python
import numpy as np

a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

# Calcular la diferencia entre los arrays
diff = a - b

# Calcular la media aritmética de la diferencia
media_diff = np.mean(diff)

print("La media aritmética de la diferencia (a-b) es:", media_diff)
```

```python
import numpy as np

# Array original
array_1 = np.array([[1.33, 4.5],
                        [30.0, 10.7],
                        [70.2, 0.5]])

# Nuevas coordenadas
array_2 = np.array([[37.1, -3.6]])

# Array con todas lasd coordenadas
array_all = np.vstack((array_1, array_2))

# Dimensiones del nuevo array
dimensiones = array_all.shape

print("El nuevo array de coordenadas es:\n", array_all)
print("Las dimensiones del nuevo array de coordenadas son:", dimensiones)
```

5) La media, mediana.. de Granada

```python
#Filtro para obtener el municipio de Granada
municipios_granada = dataset[dataset['Provincia'] == 'Granada']

# Calculando los estadisticos
media_poblacion = municipios_granada['Poblacion'].mean()
mediana_poblacion = municipios_granada['Poblacion'].median()
std_poblacion = municipios_granada['Poblacion'].std()
max_poblacion = municipios_granada['Poblacion'].max()
min_poblacion = municipios_granada['Poblacion'].min()

print(f"La población de la provincia de Granada:")
print(f"Media: {media_poblacion}")
print(f"Mediana: {mediana_poblacion}")
print(f"Desviación estándar: {std_poblacion}")
print(f"Máximo: {max_poblacion}")
print(f"Mínimo: {min_poblacion}")
```

6) Histograma de la poblacion

```python
dataset['Poblacion'] = pd.to_numeric(dataset['Poblacion'], errors='coerce')

# Agrupa el dataset por 'Provincia' y luego grafica un histograma para cada grupo
for provincia, group in dataset.groupby('Provincia'):
    plt.figure(figsize=(10, 5))
    plt.hist(group['Poblacion'], bins=15, alpha=0.7)
    plt.title(f'Histograma de Población - {provincia}')
    plt.xlabel('Población')
    plt.ylabel('Número de Municipios')
    plt.grid(True)
    plt.show()
```

7) Municipios aleatorios

```python
# Sin duplicados
dataset_sd = dataset.drop_duplicates(subset='Municipio')

# Seleccionamos aleatoriamente 50 municipios diferentes
municipios_alea = dataset_sd.sample(n=50, random_state=1)  

# Calculamos la media y la desviación estándar de la población de los municipios seleccionados aleatoriamente
media = municipios_alea['Poblacion'].mean()
std = municipios_alea['Poblacion'].std()

print(f'La media de la población de los 50 municipios seleccionados aleatoriamente es: {media}')
print(f'La desviación estándar de la población es: {std}')
```

8) Municipios mas lejanos y cercanos

```python
# Coordenadas de Madrid
madrid_lon = -3.703790
madrid_lat = 40.416775

# Analizando coordenads sin incluir a Madrid
dataset_sin_madrid = dataset[dataset['Municipio'] != 'Madrid']

# Diferencia absoluta de las coordenadas para obtener la distancia
dataset_sin_madrid['Distancia_Estimada'] = dataset_sin_madrid.apply(
    lambda row: abs(row['X'] - madrid_lon) + abs(row['Y'] - madrid_lat),
    axis=1
)

# Encuentra el municipio más cercano basado en la estimación de distancia
municipio_cercano = dataset_sin_madrid.loc[dataset_sin_madrid['Distancia_Estimada'].idxmin()]
municipio_lejano = dataset_sin_madrid.loc[dataset_sin_madrid['Distancia_Estimada'].idxmax()]
print(f"El municipio más cercano a Madrid, excluyendo a Madrid, es: {municipio_cercano['Municipio']}")
print(f"El municipio más cercano a Madrid, excluyendo a Madrid, es: {municipio_lejano['Municipio']}")
```
